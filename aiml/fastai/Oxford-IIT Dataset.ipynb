```python
IMG_HEIGHT=50
IMG_WIDTH=50
```

```python
class_names = list(map(lambda x: '_'.join(x.split('_')[:-1]), files))
```

```python
class_names
```

```python
import numpy as np
```

```python
classes = np.array(list(set(class_names)))
```

```python
len(classes)
```

```python
classes
```

```python
def get_label(path):
    parts = tf.strings.split(path , '_')
    label = tf.strings.reduce_join(parts[:-1],separator='_', axis=0)
    one_hot = label == classes
    return tf.argmax(one_hot)
```

```python
def decode_img(img):
    img = tf.io.decode_jpeg(img , channels=3)
    return tf.image.resize(img , [IMG_HEIGHT , IMG_WIDTH])
    
```

```python
def preprocess_path(file_path):
    label = get_label(file_path)
    img = tf.io.read_file(file_path)
    img = decode_img(img)
    return img , label
```

```python
AUTOTUNE = tf.data.AUTOTUNE
```

```python
val_size = int(tf.data.experimental.cardinality(list_dataset).numpy()) * 0.2
```

```python
val_size = int(val_size)
```

```python
train_ds = list_dataset.skip(val_size)
```

```python
validation_ds = list_dataset.take(val_size)
```

```python
train_ds = train_ds.map(preprocess_path , num_parallel_calls = AUTOTUNE)
validation_ds = validation_ds.map(preprocess_path , num_parallel_calls = AUTOTUNE)
```

```python
for image , label in train_ds.take(500):
    plt.imshow(image.numpy())
    print(image.numpy().shape)
    print("label: " + str(label.numpy())
```

```python
def configure_for_performance(ds):
    ds = ds.cache()
    ds = ds.shuffle(buffer_size=1000)
    ds = ds.batch(40)
    ds = ds.prefetch(buffer_size=tf.data.AUTOTUNE)
    return ds
    
```

```python
train_ds = configure_for_performance(train_ds)
```

```python
validation_ds = configure_for_performance(validation_ds)
```

```python
image_batch , label_batch = next(iter(train_ds))
```

```python
from tensorflow.keras import models
from tensorflow.keras import layers
```

```python
model = models.Sequential([
    layers.Conv2D(filters=64 , activation='relu' , input_shape=(50,50,3), kernel_size=(3,3)),
    layers.MaxPool2D(pool_size=(2,2)),
    layers.Conv2D(filters=64 , activation='relu' , kernel_size=(3,3)),
    layers.MaxPool2D(pool_size=(2,2)),
    layers.Flatten(),
    layers.Dense(100 , activation='relu'),
    layers.Dense(50 , activation='relu'),
    layers.Dense(37 , activation='softmax')
    
])
```

```python
model.compile(optimizer='adam' , loss='sparse_categorical_crossentropy', metrics=['acc'])
```

```python
model.fit(train_ds,validation_data=validation_ds,epochs=3)
```
